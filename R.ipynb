```r
x<-c(8,7,6,5,4,3,2,1)
y<-c(2.50,5.00,7.50,10.00,12.50,15.00,17.50,20.00)
```

```r
plot(x,y,col="blue", main="Bungee Jumping Demand Curve",xlab="Time Spent (in half-hours per month", ylab="Price (per half-hour)")
```

```r
relation<-lm(y~I(x))
print(relation)
summary(relation)
```

```r
x_seq <- seq(from=1,to=8, by=0.01)
new_relation<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation[i]<-predict(relation, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation),xlab="x",ylab="y")
```

```r
RSE_Percent = 9.638e-16/mean(y) * 100
```
